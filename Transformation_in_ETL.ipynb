#1 Define Data Transformation in ETL and explain why it is important.
    ->Data Transformation is the process in ETL (Extract, Transform, Load) where raw data extracted from different sources is converted into a clean, consistent, and usable format before loading it into a target system such as a data warehouse.

    Explanation:
    During transformation, data is modified by applying various operations such as:

    Cleaning (removing errors, handling missing values)

    Filtering (selecting required data)

    Aggregation (summarizing data)

    Standardization (formatting data uniformly)

    Conversion (changing data types or formats)

    Importance of Data Transformation:

    1.Improves Data Quality: Removes inconsistencies, errors, and duplicates.

    2.Ensures Consistency: Converts data into a uniform format (e.g., date formats, text cases).

    3.Enables Better Analysis: Clean and structured data helps in accurate reporting and decision-making.

    4.Integrates Multiple Sources: Combines data from different systems into a single usable format.

    5.Enhances Efficiency: Reduces processing time for analytics and machine learning tasks.

    Conclusion:
    Data Transformation is a critical step in ETL because it converts raw, unorganized data into meaningful and reliable information, making it ready for analysis and business use.

#2  List any four common activities involved in Data Cleaning.

    -> Data Cleaning is the process of identifying and correcting errors, inconsistencies, and inaccuracies in a dataset to improve its quality.

    Four common activities involved in Data Cleaning:

    1.Handling Missing Values:
    Missing data is managed by removing records, filling with mean/median/mode, or using advanced imputation techniques.

    2.Removing Duplicates:
    Duplicate records are identified and deleted to avoid redundancy and incorrect analysis.

    3.Correcting Inconsistent Data:
    Standardizing formats such as dates, text (e.g., "Male", "male", "M"), and units to maintain consistency.

    4.Outlier Detection and Treatment:
    Identifying unusual or extreme values and either correcting, transforming, or removing them based on context.

    Conclusion:
    These activities ensure the dataset is accurate, consistent, and reliable, which is essential for proper data analysis and decision-making.

#4 A dataset has missing values in the “Age” column. Suggest two techniques to handle this and explain when they should be used.
    -> Missing values in the “Age” column can be handled using the following two techniques:

    1.Mean/Median Imputation:
    In this method, missing values are replaced with the mean or median of the “Age” column.

    When to use:

    Use mean when data is normally distributed (no extreme outliers).

    Use median when data contains outliers or is skewed.

    Example:
    If average age = 25, then missing values can be filled with 25.

    2.Deletion Method (Listwise/Pairwise Deletion):
    In this method, rows with missing “Age” values are removed from the dataset.

    When to use:

    When the number of missing values is very small.

    When removing those records does not affect overall analysis.

    Example:
    If only 2–3 rows out of 1000 have missing values, they can be deleted.

    Conclusion:
    Mean/median imputation is useful when we want to retain data, while deletion is suitable when missing values are minimal and do not significantly impact the dataset.

    Each category is converted into a separate column.

    Values are represented as:

    1 → Presence of category

    0 → Absence of category


    2.It is used because:-

    One-Hot Encoding converts them into numbers without giving priority/order.

    ["Red", "Blue", "Green"]

    Red

    Blue

    Green

#7 : Explain the difference between Data Integration and Data Mapping in ETL.

    -> 1.Definition

    Data Integration
    Data Integration is the process of combining data from multiple sources (databases, files, APIs) into a single unified dataset for analysis.

    Data Mapping
    Data Mapping is the process of matching fields (columns) from the source data to the target data structure (destination table or warehouse).

    2.Key Differences:-

    Data Integration :-

    Meaning	- Combines data from multiple sources
    Purpose	- Create a unified dataset
    Focus	  - Whole data merging
    Stage in ETL	- Transformation phase
    Example	      - Combining data from CRM + ERP

    Data Mapping:-
    Meaning - Matches source fields to target fields
    Purpose - Ensure correct data placement
    Focus   - Column-to-column relationship
    Stage in ETL - Part of transformation
    Example      - Mapping “Name” → “Customer_Name”


    3.Example

    Data Integration Example:

    Data from Sales system + Customer database is merged into one dataset.


    Data Mapping Example:

    Source column “cust_name” → Target column “Customer_Name”

    Source column “dob” → Target column “Date_of_Birth”

    4.Final Explanation:-

    Data Integration focuses on bringing data together from different sources.

    Data Mapping focuses on how fields are connected between source and target.

    Both are essential steps in ETL to ensure accurate and usable data.

    Mean = 0

    .Min-Max Scaling
    It rescales data into a fixed range (usually 0 to 1)

    x' = \frac{x - x_{min}}{x_{max} - x_{min}}

    2.Key Difference (Outliers Impact)

    Min-Max Scaling

    Uses minimum and maximum values

    If outliers exist → they distort the range

    Normal data gets compressed into a small range

    Z-score Standardization

    Uses mean and standard deviation

    Less affected by extreme values

    Maintains better data distribution

    3.Example

    Suppose data:
    [10, 12, 14, 16, 100] (100 is an outlier)

    In Min-Max Scaling

    100 becomes 1

    Other values shrink close to 0 → loss of variation

    In Z-score

    Values are scaled based on mean

    Outlier remains extreme but does not compress others

    4.Final Explanation

    Min-Max Scaling is highly sensitive to outliers because it depends on min and max values.

    Z-score Standardization is more robust as it uses mean and standard deviation.

    Therefore, Z-score is preferred when datasets contain outliers.
