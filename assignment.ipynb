#1. Explain the difference between Relative, Absolute and Mixed Cell Referencing.
You have the following dataset:
Vehicle       Price
Car           500000
Bicycle       10500
Auto-Rickshaw 25000

->
  1. Relative (A1): Changes when copied.
  2. Absolute ( $A$1 ): Stays same when copied.
  3. Mixed ($A1 or A$1): Freezes column or row.

Example: If you copy =B2*1.1 (10% extra) down:
- Relative (B2): B3*1.1, B4*1.1
- Absolute ($B$2): B2*1.1, B2*1.1
- Mixed (B$2): B$2*1.1, B$2*1.1


If you want to calculate 10% extra in Column C:
- Relative: =B2*1.1 (copy down: B3*1.1, B4*1.1)
- Absolute: =$B$2*1.1 (stays B2*1.1)
- Mixed: =B$2*1.1 (freezes row, B$2*1.1)
