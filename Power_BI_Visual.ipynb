#Question 1 : Create a Card Visual to show Total Sales.
    -> Steps :-
    1.Go to Report View
    2.Select Card Visual
    3.Drag the Sales field into the card

    Result-
    The Card Visual displays the Total Sales of the dataset.

#Question 2  : Create a Donut Chart showing Total Sales by Ship Mode and which ship mode generates the highest sales?
    -> Steps:-
    1.Select Donut Chart
    2.Drag:
     .Ship Mode → Legend
     .Sales → Values
    
    Result -

    The Donut Chart shows sales distribution by Ship Mode.

    Insight

    The ship mode with the largest section in the donut chart generates the highest sales.

#Question 3 : Create a Line Chart showing the Sales Trend over time (by Month and Year). Is there any noticeable seasonal pattern in sales?
    ->Steps:-
    1.Select Line Chart
    2.Drag:
      .Order Date → X-axis
      .Sales → Y-axis
    3.Use Month-Year hierarchy
    
    Result -
    The Line Chart displays monthly sales trends over time.

    Insight

    Seasonal patterns can be identified by observing repeated peaks and drops during certain months.

#Question 4 : Create a Bar Chart to compare Sales by Category.
    -> Steps:-
    1.Select Clustered Bar Chart
    2.Drag:
      .Category → Y-axis
      .Sales → X-axis
    
    Result-
    The chart compares total sales across different categories.

#Question 5 : Create a Matrix Visual to display Sales by Month and Category and identify the month with the highest sales.
    ->Steps:-
    1.Select Matrix Visual
    2.Drag:
      .Month from Order Date → Rows
      .Category → Columns
      .Sales → Values
    
    Result -
    The Matrix displays monthly sales for each category.

    Insight

    The month with the highest total sales can be identified by the largest value in the matrix.

#Question 7 : Create a Scatter Chart to compare Discount vs. Profit by Sub-Category. Are there any outliers where high discounts resulted in negative profit?
    -> Steps:-

    2.Drag:
      .Discount → X-axis
      .Profit → Y-axis
      .Sub-Category → Legend
      .Sales → Size
    
    Result -
    The Scatter Chart compares discount levels against profit.

    Insight -
    Outliers with high discounts and negative profit indicate unprofitable discount strategies.
