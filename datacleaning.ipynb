#3.Explain the use of text functions such as TRIM , LEFT, RIGHT, MID, and CONCAT in data cleaning.

    ->- TRIM: Removes extra spaces from text, e.g., TRIM(" hello ") → "hello"
      - LEFT: Extracts characters from the start, e.g.,   LEFT("hello", 2) → "he"
      - RIGHT: Extracts characters from the end, e.g., RIGHT("hello", 2) → "lo"
      - MID: Extracts characters from the middle, e.g., MID("hello", 2, 3) → "ell"
      - CONCAT (or &): Joins text, e.g., CONCAT("hello", " world") → "hello world"

# 4. What is the role of date functions like TODAY in managing datasets?

    -> Date functions like TODAY() are super handy:
      - Track current date: TODAY() inserts the current date
      - Calculate durations: Find days between dates, e.g., =TODAY()-A1
      - Filter or categorize: Use dates to filter data, like recent records
      - Automate updates: Formulas update dynamically with current date

# 5. Apply Data Validation to restrict Quantity values to only whole numbers between 1 and 10.

Configure an input message that appears when a user selects a cell in the "Quantity" column explaining:
"Please enter a whole number between 1 and 10."

Set up an error alert message that triggers if the user enters a number less than 1 or greater than 10,
showing:
"Invalid input! The quantity must be a whole number between 1 and 10."

Write a step-by-step approach for this question.

    -> Here's a step-by-step approach to apply Data Validation:

      1. Select cells: Highlight the "Quantity" column (e.g., D2:D10).
      2. Data Validation: Go to Data > Data Validation.
      3. Settings: Choose "Whole number" and set criteria to "between 1 and 10".
      4. Input Message: Check "Show input message" and enter:
          - Title (optional): "Quantity"
          - Message: "Please enter a whole number between 1 and 10."
      5. Error Alert: Check "Show error alert" and set:
          - Style: Stop
          - Title (optional): "Invalid Input"
          - Error message: "Invalid input! The quantity must be a whole number between 1 and 10."

    Now the Quantity column restricts entries to whole numbers 1–10 with custom messages.

#8. Understanding TODAY() and NOW()

 a. What is the difference between TODAY() and NOW() in Excel? Provide an example of when you would use
 each function.

 b. If cell A1 contains the date 2025-06-10, write a formula using TODAY() to determine how many days are

 left until that date

 c. Write an Excel formula using NOW() to display the current date and time in the format MM/DD/YYYY

 HH:MM AM/PM. ́

 d. If a cell contains =TODAY(), what will happen when the worksheet is reopened the next day? Explain

 e. You want to store a static date (today’s date) in a cell without it changing every day. What keyboard

 shortcut should you use?


    -> a) TODAY()-Returns current date only and No time included
    Example:
    Excel
    =TODAY()
    Result:
    25-02-2026 (only date)
    
    NOW()- Returns current date + current time
    Example:
    Excel
    =NOW()
    Result:
    25-02-2026 11:45 PM


    B) If A1 contains date 2025-06-10, then no. of days left are-
    Formula:
    =A1 - TODAY()
    This subtracts today's date from the given date.
    If today is 25-02-2026, result will be negative (because date already passed).

    c) Display current date and time in format MM/DD/YYYY HH:MM AM/PM
    Formula:
    =TEXT(NOW(),"MM/DD/YYYY HH:MM AM/PM")
    Example Result:

    02/25/2026 11:45 PM

    d) If a cell contains =TODAY(),
    The date will automatically update to the new current date.
    Because TODAY() is a dynamic function (it changes daily).

    e) Store a static date (today’s date) that does NOT change
    Keyboard Shortcut:
    Ctrl + ;
    This inserts current date permanently (static).
    It will NOT change next day.
