#Question 1 : What is Power BI and why is it used in businesses?
    -> Power BI:-
    Power BI is a business intelligence and data visualization tool from Microsoft. It lets you connect to multiple data sources, transform raw data, build interactive dashboards, and share insights across your organization. Think Excel charts on steroids, but connected live to your databases, CRMs, and cloud apps.
    
    It is used in businesses for :-
    
    Single source of truth: Combines data from SQL, Excel, Salesforce, SAP, etc. into one dashboard. No more “which Excel is the latest” confusion.  Self-service analytics: Business users can drag-drop to build their own reports without waiting on IT. Sales managers can slice revenue by region, product, date themselves.  
    
    Real-time decision making: DirectQuery and scheduled refresh mean KPIs update hourly or daily. You spot a sales drop today, not next month.  
    
    Data storytelling: Interactive visuals, drill-downs, and AI visuals like Q&A let execs ask “why did revenue drop in Jan” in plain English and get charts back.  
    
    Cost & ecosystem: Cheaper than Tableau for most SMBs. Integrates natively with Microsoft 365, Teams, SharePoint, and Azure. Already part of many company licenses.  
    
    Overall, Companies use Power BI to turn raw data into decisions faster. It cuts manual reporting time from days to minutes and gets everyone looking at the same numbers.

#Question 2 : Name and explain the three main components of Power BI.
    -> The 3 core components are:
    
    1.Power BI Desktop -
    > What it is: Free Windows app where you build reports.  
    > What you do: Connect to data, clean it with Power Query, create relationships, write DAX measures, and design visuals.  
    > Think of it as: The workshop. This is where all the data modeling and dashboard building happens before publishing.
    
    2.Power BI Service -
    > What it is: Cloud-based SaaS at app.powerbi.com.  
    > What you do: Publish reports from Desktop, set up scheduled refresh, create dashboards, share with teams, set permissions, and view on web/mobile.
    > Think of it as: The gallery. It’s where reports go live, users consume them, and collaboration happens.
    
    3.Power BI Mobile -
    > What it is: iOS/Android/Windows apps for phones and tablets.  
    > What you do: View and interact with dashboards/reports on the go. Get push alerts when KPIs hit thresholds. Annotate and share snapshots.  
    > Think of it as: The pocket version. Lets execs check sales on their phone during meetings.
    
    They work together by:  
    Build in Desktop → Publish to Service → Consume in Service + Mobile.  
    Also, Power BI Report Server is the on-premises version of Service for companies that can’t use cloud.

#Question 3 : Explain the Power BI workflow.

    ->The standard Power BI workflow has 5 steps:
    Get → Transform → Model → Visualize → Share
    
    1.Get Data -
    Connect Power BI Desktop to your data sources. Could be Excel, SQL Server, SharePoint, Salesforce, APIs, CSVs, Azure, etc.
    Power BI uses connectors to pull the raw data in.
    
    2.Transform Data -
    Use Power Query Editor to clean the data before loading. This is your ETL layer.Examples: remove duplicates, split columns, fix date formats, filter rows, merge tables.
    
    3.Model Data -
    Define relationships between tables, e.g. Sales[Customer_ID] → Customer[Customer_ID].  
    Create DAX measures for KPIs: Total Sales = SUM(Sales[Sales_Amount]).  
    Set data types, hierarchies, categories. This is where you build the semantic model.Think of it as turning messy data into a business-friendly structure.
    
    4.Visualize Data -
    Drag fields onto the report canvas to build charts, tables, maps, slicers.
    Add interactivity: filters, drill-through, bookmarks.This step answers business questions visually: “Sales by month”, “Top 5 customers”.
     
    5.Share & Collaborate -
    Publish the .pbix file from Desktop to Power BI Service.  
    Set up scheduled refresh so data stays current.Create dashboards, set row-level security, share with users or embed in Teams/SharePoint.End users consume via web browser or Power BI Mobile.
    
    One-liner flow:  
    Source Data → Power Query → Data Model → Report → Service/Mobile → Users

#Question 4 : List any four data cleaning tasks that can be performed in Power Query.

    -> 1.Remove Duplicates -
    Delete exact duplicate rows from a table or remove duplicates based on specific columns to ensure each record is unique.  
    
    2.Change Data Types -
    Convert columns to the correct type like Text, Whole Number, Decimal, Date, or Boolean. Prevents errors in calculations and filtering.
    
    3.Replace Values / Replace Nulls -
    Find and replace incorrect values or text errors. Replace null or blank cells with a default value like 0 or "Unknown".
    
    4.Split Columns -
    Divide one column into multiple columns using a delimiter like comma, space, or fixed width. Useful for separating full names, addresses, or combined codes.

#Question 6 : Define Data View, Report View, and Model View. Explain the purpose of each view.

    -> These are the 3 main views in Power BI Desktop, found as icons on the left sidebar.
    
    1.Data View  Definition: Table-level view showing the actual data after it’s loaded from Power Query. Looks like an Excel grid.  Purpose: Inspect and validate your data. Check row counts, preview column values, create calculated columns, and confirm data types. You can’t edit source data here, only add DAX columns or measures.
    
    2.Report View  Definition: The default canvas where you build visualizations. This is where reports are designed.  Purpose: Create and arrange visuals like charts, tables, slicers, and cards. Drag fields from your model onto the canvas to answer business questions. Add interactivity like filters, drill-through, and bookmarks.
    
    3.Model View  Definition: Diagram view that shows all tables and the relationships between them.  Purpose: Manage the data model. Create or edit relationships between tables, set cardinality and cross-filter direction, hide fields, create hierarchies, and organize tables. A correct model is critical for DAX and accurate visuals.
    
    In short, Data View = See the data  Report View = Visualize the data  Model View = Connect the data.

#Question 7 : Discuss the different data sources that Power BI supports.

    -> Power BI has 150+ connectors. They fall into 4 main categories:
    1.File Sources -
    Used for flat files and local data.
    Examples: Excel .xlsx, CSV, Text, XML, JSON, PDF, Folder, SharePoint Folder.
    
    2.Database Sources -
    Connect to relational and non-relational databases. Supports Import or DirectQuery mode.
    Examples: SQL Server, MySQL, PostgreSQL, Oracle, IBM Db2, Azure SQL Database, Snowflake, Dataverse, Amazon Redshift, Google BigQuery, MongoDB.
    
    3.Online Services / Cloud Apps -
    Pull data directly from SaaS platforms using built-in APIs.
    Examples: SharePoint Online, Dynamics 365, Salesforce, Google Analytics, Adobe Analytics, Azure DevOps, GitHub, Mailchimp, QuickBooks Online.
    
    4.Other Sources -
    Covers everything else.
    Examples: Web pages via URL, OData feeds, Blank Query, R/Python scripts, Azure services like Azure Blob Storage and Azure Data Lake, Power BI Datasets and Dataflows, Microsoft Exchange.
    
    Connection Modes:  
    Most sources support Import where data is cached in Power BI, or DirectQuery where queries run live against the source. Some cloud sources also support Live Connection for Analysis Services.
    So Power BI can pull data from files on your desktop, on-prem databases, and cloud apps into one model.

#Question 8 : Split Owner Name to create two new columns as First Name and Last Name.

    -> Step 1: Open Power Query
    In Power BI Desktop, Home tab → Transform Data.
    
    Step 2: Select the column
    Click the Owner Name column header to select it.
    
    Step 3: Split Column
    Go to Home tab → Split Column → By Delimiter.  
    
    Step 4: Configure the split
    > Select delimiter: Choose Space.  
    > Split at: Select Left-most delimiter if you only want first name vs rest. For exactly two parts, Each occurrence works if names have only one space.  
    > Advanced: Under "Split into", choose Columns. Set Number of columns to 2.  
    > Click OK.
    
    Step 5: Rename the new columns
    You’ll get Owner Name.1 and Owner Name.2.  
    Double-click headers and rename to First Name and Last Name.
    
    Step 6: Apply
    Click Close & Apply to load the changes back to Power BI.
    Also, if names have middle names like "Rahul Singh Rana", using Left-most delimiter puts "Rahul" in First Name and "Singh Rana" in Last Name.
    If you need only the last word as Last Name, use Split Column → By Delimiter → Right-most delimiter instead.
