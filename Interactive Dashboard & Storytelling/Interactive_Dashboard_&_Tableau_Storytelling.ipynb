#Q1 Explain the difference between a Tableau Dashboard and a Tableau Story Point. Give two healthcare-related use cases.
    ->a)Difference:
    1.Dashboard: Single screen with multiple interactive charts/sheets placed together. All views visible at once with filters. Used for real-time monitoring.  
    2.Story Point: Sequential slides like PowerPoint.
    Each slide = 1 sheet + caption. Used for guided presentation/narrative.  
    b)Healthcare use cases:  
    1.Dashboard: Hospital admin monitoring live bed occupancy, ER wait time, patient count region-wise in one screen.  
    2.Story Point: Monthly report to board showing 1.Patient distribution → 2. Department costs → 3. Outcome analysis → 4. Recommendations.
#Q2 What are Actions in Tableau? Explain how Filter Action can help hospital management analyze patient data.
    ->Actions in Tableau: Interactive features that link sheets.When you click/select something in 1 sheet, it triggers change in other sheets.
    Filter Action for hospital management: Click on "West Region" in Region Map → All other charts on dashboard auto filter to show only West data. Helps management compare Dept cost, Doctor performance, Outcome specifically for West region without changing filters manually.

#Q7 Doctor Performance & Patient KPI Analysis

    1.New sheet → Drag Doctor Name to Rows.

                 Task 1: Doctor Performance.
