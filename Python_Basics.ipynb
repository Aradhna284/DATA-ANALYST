#Question 2 : Explain the difference between List and Tuple in Python.
    ->In Python, both List and Tuple are used to store collections of items, but they have key differences.
    A List is defined using square brackets [] and is mutable, which means we can add, remove, or change elements after it is created. Lists are generally slower and consume more memory.
    On the other hand, a Tuple is defined using round brackets () and is immutable, which means once it is created its elements cannot be modified. Tuples are faster and use less memory compared to lists. Because of immutability, tuples can also be used as keys in dictionaries, while lists cannot.
    For example: my_list = [1, 2, 3] and my_tuple = (1, 2, 3).
