#Question 1 : What is Data Loading in Power BI and why is it considered the first step of analysis?

    -> Data Loading is the process of importing or connecting data from various external sources into Power BI Desktop. It brings raw data into Power BI’s data model so it can be cleaned, transformed, and visualized.
    Key points:
    1.Connects to sources: Excel, CSV, SQL Server, SharePoint, Web, APIs, Google Sheets, etc.
    2.Two main modes:
    .Import: Creates a copy of data inside Power BI .pbix file. Fast performance.
    .DirectQuery: Leaves data at source and queries it live. Used for very large datasets.
    3.Done via Home → Get Data option in Power BI Desktop.It is considered the first step of analysis because Power BI cannot perform any analysis, transformation, or visualization without data. It establishes the foundation for all later steps.
    The Power BI workflow order:
    1.Load Data → Bring raw data in
    2.Transform Data → Clean/filter using Power Query Editor  
    3.Model Data → Create relationships, DAX measures
    4.Visualize Data → Build charts, dashboards, reports.
    Without Step 1, Steps 2-4 are impossible.
    
    Example 1: Excel File
    You have monthly sales data in Sales_Report_March2025.xlsx. In Power BI: Home → Get Data → Excel → Select file → Choose SalesSheet → LoadNow Power BI has all sales rows/columns and you can start making bar charts of revenue.
    
    Example 2: SQL Database
    School student database is stored in SQL Server.In Power BI: Home → Get Data → SQL Server → Enter server name → Select StudentMarks table → LoadThis loads all student marks into Power BI to analyze performance trends by class.

#Question 2 : Explain the difference between “Load” and “Transform Data” in Power BI.
    -> Load means bringing data from your source file or database directly into Power BI without making any changes. When you click the Load button, Power BI takes the data exactly as it is in Excel, CSV, or SQL and imports it into the data model. This is the fastest way to get data into Power BI, but if the source data has errors, empty columns, or wrong data types, those problems will also come into Power BI.
    
    Transform Data means opening the Power Query Editor before loading the data. This option lets us clean, filter, and shape the data first. We can delete unwanted columns, change data types, split or merge columns, remove duplicate rows, and fix errors. Only after we finish cleaning, we click Close & Apply and then the clean data gets loaded into Power BI. This takes more time but ensures our data is analysis-ready.The main difference is: Load brings raw, uncleaned data directly into Power BI. Transform Data cleans and prepares the data before it enters Power BI.
    
    Example 1: You have an Excel sheet of student marks with columns Name, Maths, English, Total, Remarks, and BlankColumn. If you choose Load, all columns including BlankColumn will come into Power BI. But if you choose Transform Data, you can delete BlankColumn, change Total from Text to Whole Number, and remove rows where Name is empty. Then only clean data is loaded.
    
    Example 2: A sales CSV file has OrderDate written as 12-05-2025 but Power BI reads it as Text. Using Load will keep it as Text, so you cannot use date filters or time-based charts. Using Transform Data allows you to change the column type to Date first, so later you can create monthly sales graphs.

    In short: We use Load when data is already clean and correct. We use Transform Data when data needs cleaning, formatting, or shaping before analysis. Transform Data is the recommended first step for real-world messy data.

#Question 3 : What is a Fact Table and a Dimension Table? Give examples from the dataset.
    -> A Fact Table stores the measurable, numeric data from business activities. It contains the “facts” or numbers we want to analyze like sales amount, quantity sold, marks obtained, profit, etc.
    
    Key features of Fact Table:
    1.Contains mostly numbers/measures that can be summed, averaged, counted.

    2.Has very large number of rows because it records every transaction or event.
    
    3.Contains foreign keys that link to Dimension Tables.
    4.Answers questions like “How much?”, “How many?”
    
    Dimension Table - A Dimension Table stores descriptive information about the business. It contains text, dates, categories that describe the “who, what, when, where” of the facts.
    
    Key features of Dimension Table:
    1.Contains descriptive attributes like names, categories, locations, dates.
    
    2.Has fewer rows compared to Fact Table.
    
    3.Contains a primary key that connects to the Fact Table.
    
    4.Answers questions like “Who?”, “What product?”, “Which city?”, “When?”
    
    Simple rule: Fact = Numbers to measure. Dimension = Text to filter/slice the numbers.
    
    Example 1: School Marks Dataset
    
    Suppose you have data of student exams.
    
    Fact Table: StudentMarksColumns: StudentID, SubjectID, ExamDateID, MarksObtained, TotalMarks.
    
    This table has one row for each subject a student wrote. MarksObtained = 78, 85, 92 are the measurable facts we analyze. StudentID, SubjectID are foreign keys.
    
    Dimension Table: Students
    
    Columns: StudentID, StudentName, Class, Section, Gender.
    This describes “who” got the marks. Used to filter: Show marks of Class 10-A only.
    
    Dimension Table: Subjects
    
    Columns: SubjectID, SubjectName, TeacherName
    
    This describes “what” subject the marks are for. Used to filter: Show only Maths marks.
    
    Dimension Table: Date
    Columns: ExamDateID, Date, Month, Year, Quarter.
    
    This describes “when” the exam happened.
    Used to filter: Show marks from 2024 only.

    In Power BI, the Fact Table sits in the center and Dimension Tables connect to it. When you make a chart of Total SalesAmount by City, Power BI sums SalesAmount from Fact Table and groups it using City from Dimension Table.

#Question 4 : Why is Star Schema preferred over Snowflake Schema in Power BI?
    -> In Power BI, Star Schema is preferred because it is simpler, faster, and easier for the Power BI engine to handle during analysis.
    
    A Star Schema has one central Fact Table connected directly to multiple Dimension Tables. The Dimension Tables are denormalized, meaning all related attributes stay in one table. The structure looks like a star.  

    A Snowflake Schema also has a central Fact Table, but the Dimension Tables are normalized and split into multiple related tables. This creates more joins and looks like a snowflake.
    
    Reasons Star Schema is preferred in Power BI:
    
    1.Better Performance: Power BI’s VertiPaq engine is optimized for Star Schema. Fewer table joins mean faster queries, quicker visuals, and better report loading speed. Snowflake requires more joins between small tables, which slows down DAX calculations.
    
    2.Simplicity for DAX: Writing DAX measures is easier with Star Schema because you usually do one-hop relationships from Fact to Dimension. In Snowflake, DAX has to travel through multiple related dimension tables, making formulas complex and slower.
    
    3.Easy to Understand: Star Schema is easier for users and business analysts to understand. You can clearly see which table is Fact and which are Dimensions. Snowflake becomes confusing with too many normalized tables.
    
    4.Optimized for Columnar Storage: Power BI stores data in columns. Denormalized Dimension Tables in Star Schema compress better and use less memory than many small normalized tables in Snowflake.
    
    5.Recommended by Microsoft: Microsoft’s official Power BI documentation recommends Star Schema as the best practice for data modeling.
    
    Example 1: Sales Data
    
    Snowflake Schema:
    FactSales → DimProduct → DimProductCategory → DimProductSubcategory
    Here, to find sales of “Mobile Phones”, Power BI must join 4 tables: FactSales to DimProduct to DimProductCategory to DimProductSubcategory. This is slow.
    
    Star Schema:
    FactSales → DimProduct
    In DimProduct, we keep ProductName, Category, Subcategory all in one table. Now to find sales of “Mobile Phones”, Power BI needs only 1 join: FactSales to DimProduct. Much faster.

#Question 7: Create a new column for Recovery Rate
    -> Steps:-
    1.Go to Data View
    2.Click New Column
    3.Enter this DAX formula:
    
    Recovery Rate =
    DIVIDE(
    'Covid Data'[Recovered_Cases],
    'Covid Data'[Confirmed_Cases],
    0
    ) * 100
    Result

    A new column named Recovery Rate was created successfully.

#Question 8: Create summarized table showing total confirmed cases by Country
    ->Steps :-
    1.Go to Modeling
    2.Click New Table
    3.Enter this DAX formula:
    Country_Summary =
    SUMMARIZE(
    'Covid Data',
    'Covid Data'[Country],
    "Total Confirmed Cases",
    SUM('Covid Data'[Confirmed_Cases])
     )

    Result
    
    A summarized table showing total confirmed cases by country was created successfully.
