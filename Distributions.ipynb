#Simulate 30 rolls with =RANDBETWEEN(1,6). What is the probability of rolling a 3 exactly 5 times? (Hint: Use BINOM.DIST)

    -> Parameters

    Number of trials n = 30

    Success = getting 3

    Probability of success p = 1/6

    Required successes x = 5

    Binomial Formula

    P(X=x)=\binom{n}{x} p^x (1-p)^{n-x}

    P(X=5)=\binom{30}{5}(1/6)^5(5/6)^{25}

    Excel Method (Easy)

    In Excel use:

    =BINOM.DIST(5,30,1/6,FALSE)

    Where

    5 = number of successes

    30 = number of trials

    1/6 = probability of getting 3

    FALSE = exact probability

    Result

    P(X=5) \approx 0.153

    Final Answer:
    Probability of getting 3 exactly 5 times ≈ 0.153 (15.3%)

    Simulation in Excel

    Generate 30 rolls:

    =RANDBETWEEN(1,6)

    Copy it 30 rows, then count number of 3's:

    =COUNTIF(A1:A30,3)

#4 What is the concept of standardization (z-score), and why is it important in data analysis? Explain the formula and how standardization transforms a dataset.
    -> Concept of Standardization (Z-Score)

    Standardization is the process of converting data values into a standard scale so they can be compared easily.

    It shows how far a value is from the mean in terms of standard deviations.

    This standardized value is called the Z-score.

    Formula

    Z = \frac{X - \mu}{\sigma}

    Where:

    X = data value

    μ = mean of the dataset

    σ = standard deviation

    Example

    Suppose:

    Mean = 50

    Standard deviation = 5

    Value X = 55


    Z = \frac{55-50}{5} = 1

    This means 55 is 1 standard deviation above the mean.
    

    How Standardization Transforms Data -

    After standardization:

    The new mean becomes 0

    The new standard deviation becomes 1


    So the dataset becomes a standard normal distribution.


    Standardization is important because:

    1 It allows comparison between different datasets.
    2 It helps identify outliers.
    3 Many machine learning algorithms require standardized data.
    4 It simplifies probability calculations using the normal distribution.

#5 What is Kurtosis and their type?
    -> Meaning of Kurtosis -

    Kurtosis is a statistical measure that describes the shape of a probability distribution, especially the peakedness and tail heaviness of the distribution.

    It shows how much the data has extreme values (outliers) compared to a normal distribution.

    Types of Kurtosis :-

   1. Mesokurtic

    Shape similar to normal distribution

    Moderate peak and tails

    Kurtosis = 3

    Example: Normal distribution

    2. Leptokurtic

    High peak and heavy tails
    More outliers
    Kurtosis > 3
    Data is more concentrated around the mean.


   3. Platykurtic

    Flat peak and light tails
    Fewer extreme values
    Kurtosis < 3
    Data is more spread out.

    Summary Table

    Type	Shape	Kurtosis Value

    Mesokurtic	Normal peak	= 3
    Leptokurtic	Sharp peak	> 3
    Platykurtic	Flat peak	< 3

#7 Use Excel to compute the probability of getting at least 8 successes in 15 trials with success probability 0.5 .
    -> Number of trials (n) = 15

    Probability of success (p) = 0.5

    Required: Probability of getting at least 8 successes


    "At least 8" means:

    P(X ≥ 8)
    = P(8) + P(9) + P(10) + ... + P(15)

    Instead of calculating each one manually, Excel has a formula for this.

    Method in Excel

    Excel function:
    BINOM.DIST(number_s, trials, probability_s, cumulative)
    For P(X ≥ 8) we use:
    =1 - BINOM.DIST(7,15,0.5,TRUE)

    Why 7?

    Because:
    P(X ≥ 8) = 1 − P(X ≤ 7)
    Excel calculates P(X ≤ 7) using BINOM.DIST.

    Steps in Excel:-
    1. Open Excel.
    2. Click any cell.
    3. Enter:
    =1-BINOM.DIST(7,15,0.5,TRUE)
    4. Press Enter.

    Final Result:-

    The probability will be approximately:

    0.5 (about 50%)

#8 How does log transformation help in stabilizing variance and making data more normally distributed?
    -> Log Transformation -
    Log transformation means applying the logarithm (log) function to data values.
    Example:
    Y = \log(X)

    Common logs used:

    log10
    natural log (ln)

    It Is Used in many datasets:-

    Values are highly skewed,
    Variance increases with larger values,
    Log transformation helps reduce these problems.

    3 It Stabilizes Variance by :-

    When data has large differences between values, variance becomes unstable.

    Log transformation:

    compresses large values,
    reduces spread of the data,
    This makes the variance more stable.

    It Makes Data More Normal by:-
    Many real datasets are right-skewed (long tail on the right).
    Log transformation:
    reduces the right tail,
    makes the distribution more symmetric,
    helps approximate a normal distribution.

    Example

    Original data:
    2, 5, 10, 50, 200

    After log transformation:
    0.30, 0.70, 1.00, 1.70, 2.30

    The values become closer together, reducing skewness.
