#Question 6: The population mean salary is ₹50,000 with σ = ₹5,000. If we take a sample of 100 employees, what is the standard error of the mean (SEM)?

    ->SEM = \frac{\sigma}{\sqrt{n}}
    where
    σ = population standard deviation
    n = sample size
    SEM = \frac{\sigma}{\sqrt{n}} = \frac{5000}{\sqrt{100}} = \frac{5000}{10} = 500

    SEM = 500

    The sample mean salary will typically vary by about ₹500 from the population mean.

#Question 9: A factory produces bulbs with 2% defective rate. If 5 bulbs are chosen at random, what is the probability that all are non-defective?

    ->Step 1: Convert defective rate to probability
    Defective rate = 2% = 0.02

    Probability that a bulb is non-defective:

    P(\text{non-defective}) = 1 - 0.02 = 0.98

    Step 2: Probability for 5 bulbs

    Assuming selections are independent:

    P(\text{all 5 non-defective}) = (0.98)^5

    Step 3: Calculate
    0.98^5 \approx 0.9039
