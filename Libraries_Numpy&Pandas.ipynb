```python
#Q1. Load the given retail sales dataset into a Pandas DataFrame
#and display the first 5 rows along with the dataset information to understand its structure.

import pandas as pd

# Load the dataset
df = pd.read_csv('retail_dataset .csv')

# Display first 5 rows
print("First 5 rows of the dataset:")
print(df.head())

# Display dataset information
print("\nDataset Information:")
print(df.info())
```

```python
#Q2. Inspect the dataset and identify missing values in each column.
#Display the total number of missing values present in the dataset.

# Check missing values in each column
print("Missing values in each column:")
print(df.isnull().sum())

# Total missing values
total_missing = df.isnull().sum().sum()
print("\nTotal number of missing values in dataset:", total_missing)
```

```python
# Q3. Replace missing values in Quantity and Price with their respective mean values

# 1. Fill missing values in Quantity with mean of Quantity
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())

# 2. Fill missing values in Price with mean of Price
df['Price'] = df['Price'].fillna(df['Price'].mean())

# 3. Verify - Check if missing values are now 0
print("Missing values after filling:")
print(df[['Quantity', 'Price']].isnull().sum())

```

```python
# Q4. Remove all rows where the Product Category or Region column contains missing values.

# 1. Check rows before dropping
print("Total rows before dropping:", len(df))

# 2. Drop rows where Product Category OR Region is NaN
df = df.dropna(subset=['Product Category', 'Region'])

# 3. Check rows after dropping
print("Total rows after dropping:", len(df))

# 4. Verify - make sure missing values are gone
print("\nMissing values in Product Category and Region after dropping:")
print(df[['Product Category', 'Region']].isnull().sum())
```

```python
#Q5. Create a new column called Revenue using NumPy, where:
#Revenue = Quantity × Price
#Display the updated dataset with the new column.

import numpy as np

# Creating a new column Revenue = Quantity * Price using NumPy
df['Revenue'] = np.multiply(df['Quantity'], df['Price'])

# Display the updated dataset with the new column
print("Updated dataset with Revenue column:")
print(df.head(10))  # shows first 10 rows

```

```python
#Q6. Using NumPy, calculate the total revenue generated from all transactions in the dataset.

import numpy as np

# Calculating total revenue generated from all transactions using NumPy
total_revenue = np.sum(df['Revenue'])

print("Total Revenue generated from all transactions:", round(total_revenue, 2))
```

```python
# Q7. Group dataset by Product Category and calculate total revenue per category

# Group by Product Category and sum Revenue
revenue_by_category = df.groupby('Product Category')['Revenue'].sum()

# Sort from highest to lowest revenue
revenue_by_category = revenue_by_category.sort_values(ascending=False)

print("Total Revenue generated by each Product Category:")
print(revenue_by_category)

```

```python
#Q8. Based on the revenue calculated for each product category,
#identify the top 3 categories with the highest revenue and the bottom 3 categories with the lowest revenue.


# 1. Top 3 highest revenue categories
top_3 = revenue_by_category.head(3)

# 2. Bottom 3 lowest revenue categories
bottom_3 = revenue_by_category.tail(3)

print("Top 3 Product Categories with Highest Revenue:")
print(top_3.round(2))

print("\nBottom 3 Product Categories with Lowest Revenue:")
print(bottom_3.round(2))
```

```python
#Q9. Group the dataset by Region and calculate the total revenue generated in each region.
#Identify the region with the highest revenue and the region with the lowest revenue.


# 1. Group by Region and sum Revenue
revenue_by_region = df.groupby('Region')['Revenue'].sum()

# 2. Sort from highest to lowest
revenue_by_region = revenue_by_region.sort_values(ascending=False)

print("Total Revenue generated in each Region:")
print(revenue_by_region.round(2))

# 3. Identify highest and lowest revenue region
highest_region = revenue_by_region.idxmax()
highest_value = revenue_by_region.max()

lowest_region = revenue_by_region.idxmin()
lowest_value = revenue_by_region.min()

print("\nRegion with Highest Revenue:")
print(f"{highest_region}: {round(highest_value, 2)}")

print("\nRegion with Lowest Revenue:")
print(f"{lowest_region}: {round(lowest_value, 2)}")
```

```python
#Q10. Extract the month from the Date column using Pandas and
#group the dataset by month to calculate the total monthly revenue.
#Also, calculate the mean, median, and standard deviation of the Revenue column using NumPy.

import numpy as np
import pandas as pd

# Extract month from Date column
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y', dayfirst=True)
df['Month'] = df['Date'].dt.month_name()  # creates new column with month names

# Group by Month and calculate total revenue
monthly_revenue = df.groupby('Month')['Revenue'].sum()

# Sort months in calendar order instead of alphabetical
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
monthly_revenue = monthly_revenue.reindex([m for m in month_order if m in monthly_revenue.index])

print("Total Monthly Revenue:")
print(monthly_revenue.round(2))

# Mean, Median, Std of Revenue using NumPy
revenue_values = df['Revenue'].values  # convert to numpy array

mean_rev = np.mean(revenue_values)
median_rev = np.median(revenue_values)
std_rev = np.std(revenue_values)  # population std. Use ddof=1 for sample std

print("\nRevenue Statistics using NumPy:")
print(f"Mean: {round(mean_rev, 2)}")
print(f"Median: {round(median_rev, 2)}")
print(f"Standard Deviation: {round(std_rev, 2)}")
```
