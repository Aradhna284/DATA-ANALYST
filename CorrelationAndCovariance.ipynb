#Q1. Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.
    ->Covariance:

    Covariance measures the direction of the relationship between two variables. It tells us whether the variables move together or in opposite directions.

    If covariance > 0 → variables move in the same direction

    If covariance < 0 → variables move in opposite directions

    If covariance = 0 → no linear relationship


    Formula:

    Cov(X, Y) = \frac{\sum (X - \bar{X})(Y - \bar{Y})}{n}


    Correlation:

    Correlation measures both the strength and direction of the relationship between two variables. It is a standardized form of covariance.

    Covariance only tells whether variables move together or not, but not how strongly.
    Correlation tells how strong and how consistent the relationship is.



#Q2. What does a positive, negative, and zero covariance indicate about the relationship between two variables?
    -> Covariance indicates the direction of the relationship between two variables (X and Y).

    1. Positive Covariance (Cov > 0):

    Indicates that both variables move in the same direction
    If one variable increases, the other also increases.
    If one decreases, the other also decreases.
    Example: Income and spending
    (Income ↑ → Spending ↑)

    2. Negative Covariance (Cov < 0):

    Indicates that variables move in opposite directions.
    If one variable increases, the other decreases.
    Example: Price and demand
    (Price ↑ → Demand ↓)

    3. Zero Covariance (Cov = 0):

    Indicates no linear relationship between variables.
    Changes in one variable do not affect the other.
    Example: Shoe size and intelligence
    (No relation)

    Conclusion:

    Positive → Same direction

    Negative → Opposite direction

    Zero → No linear relationship

#Q3. Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation preferred in many cases?

    ->Limitations of Covariance:
    Depends on Scale:
    The value of covariance changes when the scale of measurement changes. For example, measuring height in centimeters vs. meters will change the covariance value, making comparisons difficult.
    
    Difficult to Interpret Magnitude:
    Covariance only tells the direction of the relationship (positive or negative) but does not tell how strong the relationship is.
    
    Not Standardized:
    The range of covariance is from –∞ to +∞, so it is hard to determine whether a relationship is strong or weak.
    
    Affected by Units:
    Since covariance depends on the units of the variables, comparing covariance across different datasets is not meaningful.
    
    Correlation is Preferred because:-
    
    Standardized Measure:
    Correlation values range from –1 to +1, making it easy to interpret the strength and direction of the relationship.
    
    Unit-Free Measure:
    Correlation does not depend on the scale or units of measurement, allowing comparison between different datasets.
    
    Easier Interpretation:
    A correlation close to +1 indicates a strong positive relationship, –1 indicates a strong negative relationship, and 0 indicates no linear relationship.

#Q4. Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation coefficient. When would you prefer to use Spearman’s correlation?
    ->1. Pearson’s Correlation Coefficient (r):

    Measures linear relationship between two variables.
    Uses actual numerical values.
    Assumes data is normally distributed.
    Sensitive to outliers.
    Value ranges from –1 to +1.
    Example: Relationship between height and weight.

    2. Spearman’s Rank Correlation (ρ):

    Measures monotonic relationship (increasing or decreasing, not necessarily linear).
    Uses ranked data (positions) instead of actual values.
    Does not require normal distribution.
    Less affected by outliers.
    Value ranges from –1 to +1.
    Example: Ranking students based on marks

    Use Spearman’s correlation when:

    1. Data is in rank form (e.g., 1st, 2nd, 3rd)
    2. Relationship is non-linear but monotonic
    3. Data contains outliers
    4. Data is not normally distributed

    Conclusion:
    Pearson is used for linear and precise numerical relationships, while Spearman is preferred for ranked, non-normal, or non-linear data.

#Q5. If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you infer causation from this value? Why or why not?
    ->Interpretation of r = 0.85:

    The value 0.85 indicates a strong positive correlation between X and Y.
    
    This means:
    As X increases, Y also tends to increase.
    The relationship is strong and consistent, but not perfect.
    Example: More study time → higher marks (generally)
    Correlation does not imply causation.

    Reason:
    1.Third Variable Problem:
    A third factor (hidden variable) may affect both X and Y
    Example: Ice cream sales and drowning (both increase due to summer)

    2.Direction is Unknown:
    We cannot say whether X causes Y or Y causes X

    3.Coincidence:
    Sometimes variables are correlated by chance without any real connection.


    Conclusion:
    r = 0.85 shows a strong positive relationship.
    But it does not prove cause-and-effect relationship.

    -> Step 1: Mean of X and Y

    X̄ = \frac{2+4+6+8}{4} = \frac{20}{4} = 5

    Ȳ= \frac{3+7+5+10}{4} = \frac{25}{4} = 6.25


    Step 2: Calculate deviations and product

    X	  Y	  X−X̄	   Y−Ȳ     (X−X̄)(Y−Ȳ)

    Final Answer:

#Q7. Compute the Pearson correlation coefficient between variables A and B:
Given Data:

A	10	20	30	40	50

B	8	14	18	24	28

    -> Step 1: Mean of A and B

    Ā = \frac{10+20+30+40+50}{5} = 30

    B̄ = \frac{8+14+18+24+28}{5} = 18.4

    Step 2: Create Table

    A	  B	  A−Ā	  B−B̄	 (A−Ā)(B−B̄)	(A−Ā)²	 (B−B̄)²

    10	8	  -20	 -10.4	  208	       400	   108.16
    20	14	-10	 -4.4	    44	       100	   19.36
    30	18	 0	 -0.4	    0	         0	     0.16
    40	24	 10	  5.6	    56	       100	   31.36
    50	28	 20	  9.6	    192	       400	   92.16


    Step 3: Sum Values

    \sum (A−Ā)(B−B̄) = 500

    \sum (A−Ā)^2 = 1000

    \sum (B−B̄)^2 = 251.2


    Step 4: Apply Formula

    r = \frac{\sum (A−Ā)(B−B̄)}{\sqrt{\sum (A−Ā)^2 \cdot \sum (B−B̄)^2}}

    r = \frac{500}{\sqrt{1000 \times 251.2}}

    r = \frac{500}{\sqrt{251200}} = \frac{500}{501.2} \approx 0.998


    Final Answer:

    \boxed{r \approx 0.998}

    Interpretation (important):

    Very close to +1

    Indicates very strong positive correlation

    A and B increase almost perfectly together

#Q10. Two investment portfolios have the following returns over 5 years. Compute covariance and correlation coefficient, and interpret whether they move together.
