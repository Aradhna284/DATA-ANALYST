#Tasks

#Task 1: Category Performance Analysis
#Business Need: Management wants to know which product categories generate the highest revenue.
#Task: Create a visualization to compare total Sales by Category and identify the top-performing category.
    ->Steps:
    1.New Sheet > Drag Category to Rows.
    2.Drag Sales to Columns > It becomes SUM(Sales) bar chart.
    3.Sort descending: Click Z↓ icon on Category axis.
    4.Color by Profit: Drag Profit to Color. Red = loss, Blue = profit.
    5.Analytics pane → Drag Average Line to Table → Shows avg sales per category.
    Insight : Technology drives highest sales at $836K, 14% higher than Furniture. But Technology also shows negative profit in some sub-categories like Phones, so revenue leadership doesn’t guarantee profitability.
    
#Task 2: Sales Trend Analysis

#Business Need: The leadership team wants to understand how sales fluctuate over time to plan promotions and inventory.
#Task: Analyze Sales trends over time (monthly) to identify peak and slow periods.
    ->Steps:
    1.New Sheet > Drag Order_Date to Columns.
    2.Right-click Order_Date > Exact Date → Month/Year.
    3.Drag Sales to Rows > Line chart appears.
    4.Click highest point → Tooltip shows Dec 2017 = $116,997.
    5.Click lowest point → Tooltip shows Feb 2015 = $13,183.
    6.Analytics pane → Drag Trend Line → Linear → Shows overall growth trend.
    Insight : Sales follow strong seasonal pattern.
    December 2017 was peak at $117K due to holiday/Year-end demand. February 2015 was lowest at $13K post-holiday slump. Recommendation: Increase inventory + ad spend before December, run discount/retention campaigns in February to smooth demand.
#Task 3: Regional Contribution Analysis
#Business Need: The regional managers want clarity on how much each region contributes to overall sales.
#Task: Visualize Region-wise Sales contribution and identify the highest contributing region.
    ->Steps:
    1.New Sheet → Drag Region to Rows.
    2.Drag Sales to Columns → Horizontal bar chart.
    3.Sort Descending → West comes on top.
    4.Marks card → Label → Drag Sales + Region → Right-click Label → Quick Table Calculation → Percent of Total.
    5.Optional: Change Marks to Pie chart for % view.
    
    Insight :West region is the top contributor at 31.5% of total sales, followed closely by East at 29.5%. South is underperforming at only 17.3%. Strategy: Double down on West/East for growth, investigate pricing, demand, or competition issues in South region.

#Task 4: Profitability Investigation

#Business Need:The finance team suspects that some high-sales orders may be reducing overall profit.

#Task:Analyze the relationship between Sales and Profit to identify loss-making or low-profit orders.
    ->Steps:
    1.New Sheet → Drag Sales to Columns, Profit to Rows → Scatter plot.
    2.Drag Order_ID to Detail on Marks card.
    3.Marks card → Color → Drag Profit → Red = negative profit.
    4.Drag filter Profit < 0 to Filters shelf → Shows only loss-making orders.
    5.Right-click → Sort → Profit Ascending → Worst orders on bottom.
    
    Insight: Finance concern is valid. 12% of high-value orders are loss-making, mainly in Technology/Phones due to discounts >30%. Management should set minimum profit margin rules and review discount policy for high-sales orders.
#Task 5: Business Growth Evaluation

#Business Need:Senior management wants to assess whether the business is growing consistently.
#Task:Visualize cumulative Sales growth over time to evaluate overall performance.
    ->  Steps:
    1.New Sheet → Drag Order_Date to Columns → Month/Year continuous.
    2.Drag Sales to Rows → Line chart.
    3.Right-click Sales pill on Rows → Quick Table Calculation → Running Total.
    4.Analytics pane → Drag Trend Line → Linear.
    5.Click trend line → Describe Trend Line → Check slope.
    
    Insight: Business is growing consistently and accelerating. Cumulative sales show 3x growth from 2015 to 2017 with positive trend line. No stagnation periods. Management should invest in capacity/scaling to support continued growth.

#Q9. Chart Selection Justification

#Explain why different charts were used for different business questions in this analysis.
    ->Different charts were chosen based on the business question type:
    1.Bar Chart for Category/Region Analysis: Used in Task 1 & 3 because we needed to compare discrete categories/regions against sales. Bars make it easy to rank "West > East > Central" and "Technology > Furniture" at a glance. Length encodes value clearly for comparison questions.
    2.Line Chart with Running Total for Trend Analysis: Used in Task 5 & 8 because CEO asked "is business growing consistently". Lines show continuous change over time better than bars. Running total shows cumulative growth trajectory and acceleration, not just monthly spikes.
    3.Scatter Plot for Profitability Investigation: Used in Task 4 because question was "relationship between sales and profit". Scatter shows correlation + outliers. We plotted Sales vs Profit and colored by profit to instantly spot high-sales but loss-making orders like "Phones". Size can add Quantity for 3rd dimension.
    4.Heat Map for Category x Region: Used in Task 6 because we needed 2-dimensional comparison. Color intensity shows which Region-Category combo is strong/weak. "West-Technology" dark blue vs "South-Furniture" red tells ops where to focus.
    5.KPI Cards for Executive Summary: CEO needs numbers, not charts. KPI cards show Total Sales, Total Profit, Profit % in big font for instant decision-making.
