#4. Median in Real Life (Medium)
A car dealer notes the prices of used cars: [$8,000, $9,500, $10,200, $11,000, $50,000].
Why is the median a better measure than the mean in this case? Calculate the median.

    -> Step 1: Arrange the numbers (already in order)
    8,000
    9,500
    10,200
    11,000
    50,000
    There are 5 numbers.
    Step 2: Find the Median
    Formula: If the number of values is odd :-
    \text{Median} = \text{Middle value}
    The middle value (3rd number) is:
    Median = $10,200

    Step 3: Why is Median Better Than Mean Here?
    Most car prices are around $8,000–$11,000
    But one car costs $50,000 (very high)
    That $50,000 is an outlier (extreme value).
    
    If we calculate the mean:
    Sum = 8,000 + 9,500 + 10,200 + 11,000 + 50,000
    = 88,700
    Mean = 88,700 ÷ 5
    = $17,740

    But:
    $17,740 does NOT represent most car prices
    Most cars are around $10,000.

    Conclusion
    Median = $10,200
    Median is better because it is not affected by the very high price ($50.It gives a more realistic idea of typical car prices.

#5. Dispersion Introduction (Medium)
A student times how long it takes to finish a puzzle each day: [25, 30, 27, 35, 40].
What does the range tell us about the variation in the student’s puzzle-solving time?

    ->Range = Highest value − Lowest value
    It shows how much the data spreads out.
    Lowest time = 25 minutes
    Highest time = 40 minutes
    Now calculating the Range:-
    {Range} = 40 - 25 = 15
    
    Range = 15 minutes
    This mean:-
    The student’s solving time varies by 15 minutes.
    Some days the student finishes quickly (25 minutes).
    Some days it takes much longer (40 minutes).
    So, the performance is not completely consistent.

#9. Combining Measures (Hard)
A family records their monthly electricity usage (in kWh): [400, 420, 390, 450, 410].
Find the mean and standard deviation. What do these values together tell you about the family’s energy use
pattern?

    ->Step 1: Find the Mean
    {Mean} = \frac{400 + 420 + 390 + 450 + 410}{5}
    Sum = 2070
    {Mean} = 2070 \div 5 = 414
    Mean = 414 kWh
    Step 2: Find Standard Deviation
    Step A: Find deviations from mean (414)
    
    Value	Deviation	Square

    400	-14	196
    420	6	36
    390	-24	576
    450	36	1296
    410	-4	16

    Sum of squares =
    196 + 36 + 576 + 1296 + 16 = 2120
    
    Step B: Variance
    {Variance} = 2120 \div 5 = 424

    Step C: Standard Deviation

    \sqrt{424} \approx 20.6

    Standard Deviation ≈ 20.6 kWh

    Final Interpretation

    Mean = 414 kWh → Average monthly usage

    Standard Deviation ≈ 21 kWh → Usage varies about 21 units from average

    This means:

    1.Electricity usage is fairly consistent.
    2.No extreme spikes.
    3.Bills will usually be close to the average
    4.The mean shows the average monthly usage, while the standard deviation shows moderate variation. The family’s electricity consumption is relatively stable.

#10.Practical Application (Hard)
A basketball player’s points in 8 games are recorded: [15, 18, 20, 22, 25, 17, 19, 21].
Find the mean, median, mode, range, and standard deviation. What insights can these measures provide
about the player’s scoring performance?

    ->Step 1: Arrange Data in Order

    15, 17, 18, 19, 20, 21, 22, 25

    Step 2: Mean (Average)

    Sum = 15 + 18 + 20 + 22 + 25 + 17 + 19 + 21
    = 157

    Mean = 157 ÷ 8
    = 19.63

    Mean ≈ 19.6 points

    Step 3: Median

    There are 8 numbers (even count).

    Median = Average of 4th and 5th values
    = (19 + 20) ÷ 2
    = 19.5
    
    Step 4: Mode

    No number repeats. No mode
    Step 5: Range

    Range = Highest − Lowest
    = 25 − 15
    = 10


    Step 6: Standard Deviation (Approx.)

    After calculation:

    Standard Deviation ≈ 3.1

    (This shows how much scores vary from the mean.)


    Final Results Summary

    Mean ≈ 19.6

    Median = 19.5

    Mode = None

    Range = 10

    Standard Deviation ≈ 3.1


    What Does This Tell Us?

    Player scores around 19–20 points per game

    Scores are fairly consistent

    Variation is small (SD ≈ 3.1)
    No extreme outliers
    This player is stable and reliable, not highly inconsistent.
