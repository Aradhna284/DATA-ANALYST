#1 What is the key difference between publishing a dashboard on Tableau Public versus Tableau Server/Cloud in terms of accessibility and data security?
    ->
    1.Tableau Public: Anyone on the internet can view the dashboard, so it should only be used for non-confidential data.
    2.Tableau Server/Cloud: Only authorized users with the required permissions can access the dashboard, making it suitable for business or sensitive data.
    Example:
    A company's annual sales report should be published on Tableau Server/Cloud because only employees should see it. A dashboard showing public COVID-19 statistics can be published on Tableau Public.

#2 When publishing a Tableau dashboard, what happens to a live connection compared to an extract? How does this affect data refresh and performance?
    ->
    1.Live Connection: The dashboard always displays the latest data directly from the database. It depends on the database speed, so performance may be slower.
    2.Extract: Tableau stores a copy of the data. Dashboards load faster, but the data must be refreshed regularly to stay up to date.
    Example:
    A stock market dashboard should use a live connection because prices change continuously. A monthly sales report can use an extract since data changes less frequently.

#3 In Tableau Server or Cloud (conceptually), which feature allows administrators to control which users can view, interact with, or edit a published dashboard?
    -> The Permissions feature allows administrators to control user access. Permissions can be assigned to individual users or groups.
    Example:
    Managers can be given permission to edit a dashboard, while employees can only view it.

#4 If you want users to view a dashboard but not download the underlying data, which permission setting would you adjust in Tableau Server?
    -> Disable the Download Data (or Download Full Data) permission while allowing View access.
    Example:
    Employees can view sales charts but cannot download customer information or raw sales data.

#5 Name two methods by which a Tableau dashboard can be embedded into a web page or website.
    ->
    1.Using the Embed Code (HTML iframe) provided by Tableau.
    2.Using the Tableau JavaScript Embedding API for interactive integration.
    Example:
    A company can embed its sales dashboard into its internal employee portal using an iframe.

#6 List two best practices you should follow when preparing a dashboard for public sharing to ensure data security and privacy.
    ->
    1.Remove or hide sensitive information such as customer names, phone numbers, or financial details.
    2.Review permissions and ensure only intended data is visible before publishing.
    Example:
    Instead of showing individual employee salaries, publish only department-wise salary totals.

#7 After publishing a dashboard, users report a “Data not found” error. What is the most common cause related to data source settings?
    -> The most common cause is that the data source is missing, moved, or the connection is broken. Incorrect database credentials or file paths can also cause this error.
    Example:
    If an Excel file used by the dashboard is moved to another folder, Tableau cannot locate it and displays a "Data not found" error.

#8 When embedding a Tableau dashboard into a secure internal portal, what two additional security measures should be implemented to ensure only authorized users can access the embedded content?
    ->
    1.Use user authentication (login or Single Sign-On) to verify users.
    2.Apply Tableau permissions or Row-Level Security (RLS) so users only see the data they are allowed to access.
    Example:
    In a company HR portal, employees must log in first, and each manager can only view data for their own department.
