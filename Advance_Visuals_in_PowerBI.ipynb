#1 What is the difference between a Basic Filter, Advanced Filter, and a Slicer in Power BI?
    ->A Basic Filter lets you check specific values from a list, like picking 2023 or North in the Filters pane.
    An Advanced Filter uses rules, like Sales > 1000 or Date is after 2023-01-01.
    
    A Slicer is a visual on the report page that users click, like a Region dropdown — when you click South, all charts update to show only South data.
    
    Key difference: Basic/Advanced = set by you in the Filters pane. Slicer = put on canvas for users to click.

#4 What is a Visual-Level Filter, and how does it differ from Page-Level and Report-Level Filters?
    -> The three filter levels in Power BI control how broadly a filter applies.
    A Visual-Level Filter affects only one specific chart or table. For example, you can set your pie chart to show only Category = Electronic while a table next to it still shows both Electronic and Accessori.
    A Page-Level Filter applies to every visual on one report page. If you add Country = India as a page-level filter on Page 1, all charts on that page show India only, but Page 2 stays unaffected.
    A Report-Level Filter applies to the entire .pbix file across all pages. Setting Country = India at report-level means Dubai data is hidden from every visual, every page, for everyone.
    Key idea: Visual = one chart. Page = one tab. Report = whole file.

#10 Explain the difference between Drill-Down and Drill-Through in Power BI using a real-world example.
    -> Drill-Down and Drill-Through both help you see more detail, but they work differently.
    
    Drill-Down lets you zoom from summary to detail within the same chart using the ↑↓ arrows. For example, in a bar chart of Country sales, you can drill down to see the Region breakdown without leaving the page. It’s like using Google Maps to go from Country → State → City on one screen.
    
    Drill-Through instead jumps you to a different report page that’s built for that item. For example, right-click Laptop in a summary table → Drill through to a Product Details page that shows Laptop sales by Region, Month, and Customer_Name. It’s like clicking a Wikipedia link to open a dedicated article.
    Rule of thumb: Use Drill-Down for quick exploration in the same visual. Use Drill-Through when you need a detailed page with more visuals and fields for that specific item.

#13 What is the “Keep All Filters” option in Drill-Through? How does it affect user interaction
    -> The "Keep All Filters" option controls what context carries over when you drill through to another page.
    When Keep All Filters = On, the target page keeps all the filters from your source page.
    For example, if Page 1 is filtered to India + 2023 and you drill through on Laptop, the target page will show only Laptop sales for India in 2023.
    When Keep All Filters = Off, the target page ignores your source page filters and only applies the item you clicked. So the same drill-through would show Laptop sales for all countries and all years.
    
    Rule of thumb: Turn it On to keep your analysis focused and predictable. Turn it Off if you want to see the broader picture for that one item.
