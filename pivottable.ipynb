# 1. What is the purpose of a PivotTable in Excel?

    -> A Pivot Table helps summarize and analyze large data sets by creating custom views, like totals, averages, or counts, without altering the original data. It's a super flexible summary tool which helps in spotting trends, comparing data and in making reports more readable.

# 2. Create a PivotTable to show Total Sales by Region.

    -> Steps-
      1. Make a column of total sales.
      2. In F2, type =D2*E2 to calculate total sales.
      3. Drag down till F21.
      4. In Inser tab, select Pivot table then select the entire table.
      5. Pivot table will be created.
      6. In the pivot field list, select Region and Total Sales.
      7. Pivot table will show Total Sales by Region.

#3. Apply Conditional Formatting to highlight sales greater than 400.

# 5. Create a PivotChart showing Monthly sales trend.
    -> Steps:
    1. Go to Insert, and select Pivot table with chart.
    2. Select table range and in the existing sheet and click OK
    3. Pivot table will be created, now select date and total sales column
    4. Right click any date inside Pivot Table and click Group ,then select Months and Years, and click Ok.
    5. Now it shows Monthly Sales Total.
    6. Pivot Chart of monthly sales will be created along this by default.

# 7. Why are Dashboards important for decision-making in business analysis?

    
    -> Dashboards give a quick visual snapshot of key metrics, helping businesses make informed decisions faster by highlighting trends, issues, and opportunities in real-time.
    Dashboards show important business info at a glance, helping us make quick, smart decisions .
    Think of dashboards like a car's dashboard - we see speed, fuel, and warnings instantly. Business dashboards show key data like sales, traffic, or profits, helping us steer decisions.
