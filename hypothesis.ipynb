#Question 1: What is a null hypothesis (H₀) and why is it important in hypothesis testing?
    -> A null hypothesis (H₀) is a default assumption that there is no effect, no difference, or no relationship between variables.
    In simple words:
    -It assumes nothing special is happening.
    Example:-
    Suppose a company claims a new medicine reduces fever faster.
    Null Hypothesis (H₀):
    The new medicine does not reduce fever faster than the old one.
    Alternative Hypothesis (H₁):
    The new medicine does reduce fever faster.
    So H₀ always represents the status quo / no change.
    The null hypothesis is important because it gives us a baseline to test against.
    Start by assuming H₀ is true
    Collect data
    Use statistical tests
    Decide whether to reject H₀ or fail to reject H₀
    Two Possible Results
    1.Reject H₀
    Evidence suggests the effect exists.
    2.Fail to Reject H₀
    Evidence is not strong enough to prove the effect.
    3.Simple Real-Life Example
    Suppose a school says:
    "Online classes improve student performance."
    H₀: Online classes do not improve performance.
    H₁: Online classes improve performance.
    After collecting student scores and running tests, we decide whether to reject H₀.

#Question 2: What does the significance level (α) represent in hypothesis testing?

    -> The significance level (α) is the probability of rejecting the null hypothesis (H₀) when it is actually true.
    
    During hypothesis testing:

    1. Choose α (significance level)


    2. Calculate the p-value from sample data


    3. Compare p-value with α



    If p-value ≤ α → Reject H₀

    If p-value > α → Fail to reject H₀

#Question 3: Differentiate between Type I and Type II errors.
    -> A Type I error occurs when a true null hypothesis is rejected, while a Type II error occurs when a false null hypothesis is not rejected. Type I error probability is α, and Type II error probability is β.

#Question 4: Explain the difference between a one-tailed and two-tailed test. Give an example of each.

    ->One-Tailed Test

    A one-tailed test checks if a parameter is either greater than or less than a specific value (only one direction).
    -The rejection region is on one side (one tail) of the distribution.
    Types-
    Right-tailed test: checks if value is greater than a number.
    Left-tailed test: checks if value is less than a number
    Example:-
    A company claims that the average battery life is more than 10 hours.
    H₀: μ ≤ 10
    H₁: μ > 10

    Here we only check the right side, so it is a one-tailed test.

    2.Two-Tailed Test

    A two-tailed test checks if a parameter is different from a specific value (either greater or smaller).

    -The rejection region is on both sides (two tails) of the distribution.
    Example
    A company claims the average battery life is 10 hours.
    We test if it is different from 10.

    H₀: μ = 10

    H₁: μ ≠ 10

    Here we check both directions, so it is a two-tailed test.
    
    3.Key Difference:-
    -One-Tailed Test	            
    *Tests one direction (greater OR less)	  
    *Rejection region in one tail	Rejection region in two tails.
    Example: μ > 10
    -Two-Tailed Test
    *Tests both directions.
    *Rejection region in two tails.
    *Example: μ ≠ 10

#Question 5: A company claims that the average time to resolve a customer complaint is 10 minutes.
A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes.

At α = 0.05, test the claim.

    ->This is a hypothesis testing problem using a t-test because the sample size is small (n < 30).

    Step-by-Step Solution

    1.State the Hypotheses

    Null Hypothesis (H₀): μ = 10 minutes

    Alternative Hypothesis (H₁): μ ≠ 10 minutes


    This is a two-tailed test.

    2.Given Data

    Sample mean (x̄) = 12

    Population mean (μ) = 10

    Standard deviation (s) = 3

    Sample size (n) = 9

    Significance level (α) = 0.05

    3.Test Statistic Formula (t-test)

    t = \frac{\bar{x} - \mu}{s/\sqrt{n}}

    4.Calculate the t-value

    t = \frac{12 - 10}{3/\sqrt{9}}

    t = \frac{2}{3/3}

    t = \frac{2}{1}

    t = 2

    5.Find Critical Value

    Degrees of freedom:

    df = n - 1 = 9 - 1 = 8

    From the t-table at α = 0.05 (two-tailed):
    t_{critical} ≈ ±2.306

    6.Decision Rule

    If |t| > 2.306 → Reject H₀
    If |t| ≤ 2.306 → Fail to reject H₀
    Here:
    |2|< 2.306
    So we fail to reject H₀.
    So,there is not enough statistical evidence at the 0.05 significance level to reject the company's claim that the average complaint resolution time is 10 minutes.

#Question 6: When should you use a Z-test instead of a t-test?

    ->When to Use a Z-Test

    A Z-test is used when the sample size is large or the population standard deviation is known.

    Main Conditions-

    1.Sample size is large

    Usually n ≥ 30

    2.Population standard deviation (σ) is known

    3.Data is normally distributed or approximately normal.

    When to Use a t-Test (for comparison)

    A t-test is used when:

    Sample size is small (n < 30)

    Population standard deviation is unknown

    We use sample standard deviation (s) instead.

    Example:-

    Z-test Example -

    A factory claims the average weight of a product is 500g.
    You take a sample of 50 items and the population standard deviation is known.
    Use a Z-test.

    t-test Example :-

    A teacher wants to test the average marks of 10 students and population standard deviation is unknown.
    Use a t-test.
    
    Key Difference
    
    Z-Test -
    *Large sample (n ≥ 30)
    *Population σ known
    *Uses Z distribution

    t-Test -
    *Small sample (n < 30)
    *Population σ unknown
    *Uses t distribution

#Question 7: The productivity of 6 employees was measured before and after a training program.
Employee    Before    After
1           50        55
2           60        65
3           58        59
4           55        58
5           62        63
6           56        59

    -> This is a paired t-test because the same employees are measured twice (before and after).

    Step 1. Hypotheses

    H₀: Training does not improve productivity (μd = 0)

    H₁: Training improves productivity (μd > 0)


    Step 2. Calculate Differences (After − Before)

    Employee	Before	After	Difference (d)

          1	    50	    55	    5
          2	    60	    65	    5
          3	    58	    59	    1
          4	    55	    58	    3
          5	    62	    63	    1
          6	    56	    59	    3


    Step 3. Mean Difference

    \bar d = \frac{5+5+1+3+1+3}{6}

    \bar d = \frac{18}{6} = 3

    Step 4. Standard Deviation of Differences

    Differences: 5, 5, 1, 3, 1, 3
    s_d ≈ 1.79

    Step 5. t-Statistic

    Formula:

    t = \frac{\bar d}{s_d / \sqrt{n}}

    t = \frac{3}{1.79/\sqrt6}

    t ≈ 4.10

    Step 6. Critical Value

    Degrees of freedom:

    df = n-1 = 6-1 = 5

    From t-table at α = 0.05 (one-tailed)

    t_{critical} ≈ 2.015

    Step 7-

    4.10 > 2.015

    So we reject H₀.
    Therefore,At α = 0.05, there is significant evidence that the training program improves employee productivity.


#Question 8: A company wants to test if product preference is independent of gender.
Gender    Product A   Product B   Total
Male           30          20        50
Female         10          40        50
Total          40          60        100

At α = 0.05, test independence

    -> This problem uses a Chi-Square Test of Independence.

    Step 1. State the Hypotheses

    H₀ (Null Hypothesis): Product preference is independent of gender.
    
    H₁ (Alternative Hypothesis): Product preference depends on gender.


    Step 2. Observed Frequencies (Given Table)

    Gender	Product A	   Product B	  Total

    Male	     30	        20	         50
    Female	   10	        40	         50
    Total	     40	        60	         100

    Step 3. Expected Frequencies

    Formula:
    E = \frac{(Row\ Total × Column\ Total)}{Grand\ Total}
    Male – Product A

    E = \frac{50 × 40}{100} = 20
    Male – Product B

    E = \frac{50 × 60}{100} = 30
    Female – Product A

    E = \frac{50 × 40}{100} = 20
    Female – Product B

    E = \frac{50 × 60}{100} = 30

    Expected Table:

    Gender	Product A	   Product B

    Male	       20	       30
    Female	     20	       30

    Step 4. Chi-Square Formula

    \chi^2 = \sum \frac{(O - E)^2}{E}

    Where
    O = Observed value
    E = Expected value

    Step 5. Calculate χ²

    Cell	      O	    E	    (O−E)²/E

    Male A	    30	  20	    5
    Male B	    20	  30	    3.33
    Female A	  10	  20	    5
    Female B	  40	  30	    3.33


    \chi^2 = 5 + 3.33 + 5 + 3.33

    \chi^2 ≈ 16.66

    Step 6. Degrees of Freedom

    df = (rows-1)(columns-1)

    df = (2-1)(2-1) = 1

    From the Chi-square table at α = 0.05:

    \chi^2_{critical} = 3.84

    Step 7. Decision

    16.66 > 3.84

    So we reject H₀.
    
    So,there is significant evidence at α = 0.05 that product preference depends on gender.
    Therefore, product preference is not independent of gender.
