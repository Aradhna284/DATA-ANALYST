#Question 1 : Create a Heatmap to analyze Treatment Cost by Department and Region.
    ->Steps:
    1.Drag Department to Columns, Region to Rows.
    2.Drag Treatment Cost to Color + Size + Label.
    3.Aggregation = SUM.
    4.Click Color legend → Edit colors → Diverging Red-Blue for better contrast.
    Insight: Darker/larger squares = higher treatment cost.
    Will use this to spot expensive dept+region combos.

#Question 2 : Create a Highlight Table showing Satisfaction Score by Department.
    ->Steps:
    1.Drag Department to Columns.
    2.Drag Satisfaction Score to Text + Color.
    3.Aggregation = AVG.
    4.Marks card → Change to Square.
    5.Analysis → Table Layout → Turn off "Show Row Grand Totals"
    Insight: Color intensity shows dept performance. AVG >4 = good, <3 = needs action.

#Question 3 : Create a Bump Chart to show ranking of Departments based on Treatment Cost over months.
    -> Steps:
    1.Drag Month to Columns. Right click → Discrete.
    2.Drag Department to Color + Label.
    3.Create calc field Rank_Cost: RANK(SUM([Treatment Cost]))
    4.Drag Rank_Cost to Rows → Compute Using = Department.
    5.Marks = Line + Circle. Sort Month ascending.
    Insight: Lines crossing = rank changes. Top line = highest cost dept that month.

#Question 4 : Create a Butterfly Chart comparing Male vs Female patient count by Department.
    -> Steps:
    1.Drag Department to RowsDrag Gender to Columns.
    2.Drag Patient ID to Columns → CNT + Label.
    3.Right click Male bar → Dual Axis → Synchronize Axis.
    4.For Male: Compute = -CNT[Patient ID] to flip left.
    5.Marks = Bar for both.
    Insight: Mirror shape shows gender split per dept. Left = Male, Right = Female.

#Question 5 :  Create a Dual Axis Chart to compare Average Age and Treatment Cost by Department.
    ->Steps:
    1.Drag Department to Columns.
    2.Drag Age to Rows → AVG + Label.
    3.Drag Treatment Cost to Rows → SUM.
    4.Right click Treatment Cost pill → Dual Axis.
    5.Right click right axis → Synchronize Axis.
    6.Marks: Age = Circle, Treatment Cost = Bar.
    Insight: See if older patients correlate with higher cost per dept.

#Question 6 : Create a Sunburst Chart showing Department → Diagnosis hierarchy by Treatment Cost.
    ->Tableau has no native Sunburst.
    Use Donut method:
    Steps:
    1.Create calc Path = [Department] + "/" + [Diagnosis].
    2.Create calc Level1 = SPLIT([Path], "/", 1).
    3.Create calc Level2 = SPLIT([Path], "/", 2).
    4.Drag Level1 to Columns → Size, then Level2 to Columns → Size.
    5.Drag Treatment Cost to Color + Size → SUM.
    6.Marks = Pie → Show Mark Labels.
    Insight: Inner ring = Dept share, outer ring = Diagnosis share within dept by cost.

#Question 7 : Create a Dual Axis Chart comparing Treatment Cost and Satisfaction Score by Department.
    -> Steps:
    1.Drag Department to Columns.
    2.Drag Treatment Cost to Rows → SUM.
    3.Drag Satisfaction Score to Rows → AVG.
    4.Right click Satisfaction Score → Dual Axis → Synchronize.
    5.Marks: Cost = Bar, Satisfaction = Line with Circle.
    6.Color code Satisfaction line Red/Green for <3 vs >4
    Insight: Check if higher cost = higher satisfaction. Negative correlation = problem.
