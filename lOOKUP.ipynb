#4.Find the Department of Employee ID = E103. Also explain which LOOKUP function is used and why?
EmpID   Name   Department
E101    Raju   IT
E102    Riya   HR
E103    Roshni Finance
E104    Sejal  Sales

    ->=VLOOKUP("E103",A2:C5,3,FALSE )
image.png

We used VLOOKUP as the data is in vertical format, however the best lookup function will be Xlookup as it works in both sides and also there is no need to count column number.
