#1 What are the most common reasons for missing data in ETL pipelines?
    ->Missing data in ETL (Extract, Transform, Load) pipelines can occur due to multiple reasons at different stages:
    1.Data Source Issues (Extraction Stage)
    Data may not be collected properly from source systems (e.g., user skips form fields).
    System errors or sensor failures can lead to incomplete records.

    2.Data Integration Problems
    When combining data from multiple sources, some fields may not match or may be absent in one source.
    
    Schema mismatches can result in null values.

    3.Data Corruption or Loss
    Network failures, system crashes, or file corruption during extraction or transfer can cause missing values.

    4.Transformation Errors
    Incorrect data cleaning or transformation logic may accidentally remove or nullify values.
    
    Type conversion errors (e.g., string to integer) can produce missing values.

    5.Filtering and Business Rules
    Certain ETL rules may intentionally drop data that does not meet conditions, leading to gaps.

    6.Delayed or Incomplete Data Loads
    Data may arrive late or partially during batch processing, causing temporary missing values.

    7.Human Errors
    Manual data entry mistakes or incorrect configurations in ETL pipelines can introduce missing data.


    Conclusion

    Missing data in ETL pipelines is usually caused by issues in data collection, integration, transformation, or system failures. Identifying the root cause is important before deciding how to handle missing values.


#2 Why is blindly deleting rows with missing values considered a bad practice in ETL?
    -> Blindly deleting rows with missing values is considered a bad practice because:

    1.Loss of Important Data

    Deleting rows may remove valuable information, reducing the dataset size significantly.

    This can lead to incomplete analysis.


    2.Introduction of Bias

    If missing data is not random (e.g., income missing mostly for low-income users), deleting such rows can bias results.

    This leads to incorrect conclusions.


    3.Distortion of Insights

    Removing too many records can affect statistical measures like mean, trends, and distributions.

    Business decisions based on such data may be wrong.


    4.Ignoring Root Cause

    Simply deleting data does not address why the data is missing.

    The underlying issue in the ETL pipeline remains unresolved.


    5.Not Always Necessary

    Many techniques (like imputation, forward fill, or using default values) can handle missing data better without losing information.

    Conclusion

    Blind deletion should be avoided because it leads to data loss, bias, and inaccurate insights. Instead, appropriate handling methods should be chosen based on the nature of missing data.

#3 Explain the difference between:Listwise deletion, Column deletion.
    ->1. Listwise Deletion (Row Deletion)

    In this method, entire rows are removed if they contain any missing values.

    It reduces the number of records in the dataset.

    When appropriate:

    When the dataset is large and only a small number of rows have missing values.

    When missing data is random and removing rows will not affect overall analysis.


    2.Column Deletion

    In this method, entire columns are removed if they contain many missing values.

    It reduces the number of features (variables) in the dataset.


    When appropriate:

    When a column has a high percentage of missing values (e.g., 70–80% missing).

    When the column is not important for analysis or modeling.

    Key Difference

    Listwise deletion removes rows (records).

    Column deletion removes columns (features).

    Conclusion

    Both methods are simple ways to handle missing data, but they should be used carefully depending on data size, importance, and the amount of missing values.
#4 Why is median imputation preferred over mean imputation for skewed data such as income?
    -> Median imputation is preferred over mean imputation for skewed data because:

    1.Less Affected by Outliers

    In skewed data (like income), there are extreme values (very high or very low).

    The mean gets influenced by these outliers, while the median remains stable.

    2.Better Representation of Central Tendency

    The median represents the middle value, which reflects the typical data point more accurately in skewed distributions.

    Mean can give a misleading average.

    3.Maintains Data Distribution

    Using median helps preserve the original shape of the data.

    Mean imputation can distort the distribution by pulling values toward extremes.

    4.More Realistic Values

    Median provides a more realistic and reliable estimate for missing values in datasets like income, where values are not evenly distributed.

    Conclusion

    For skewed data, median imputation is preferred because it is robust to outliers, preserves data distribution, and provides more accurate and meaningful results.
#5.What is forward fill and in what type of dataset is it most useful?
    ->Forward Fill:-
    Forward fill is a data imputation technique where missing values are replaced with the last known (previous) valid value in the dataset.
    It is also called “Last Observation Carried Forward (LOCF)”.

    It is most useful in:-

    1.Time-Series Data

    Data where values are recorded over time (e.g., daily sales, stock prices, sensor readings).
    It assumes that the previous value is still valid until a new value appears.

    2.Sequential or Ordered Data

    When the order of data matters and values change gradually.

    Example:-

    If sales data is:

    Day 1: 100

    Day 2: — (missing)

    Day 3: 120


    After forward fill:

    Day 2 will become 100 (copied from Day 1)

    Conclusion

    Forward fill is a simple and effective method for handling missing values in time-series or sequential datasets where previous values are meaningful.

#6. Why should flagging missing values be done before imputation in an ETL workflow?
    ->Flagging missing values before imputation is important because it preserves information about the original data and improves analytical accuracy.

    Explanation:

    1.Retains Original Information
    When we impute (fill) missing values, the fact that the data was originally missing gets lost. By creating a flag column (e.g., is_missing_income = 1/0), we keep track of which values were missing.


    2.Captures Hidden Patterns
    Missing data itself may carry meaning. For example, customers who do not report income might belong to a specific segment. Flagging helps models and analysts detect such patterns.


    3.Improves Model Performance
    Machine learning models can use the flag as an additional feature, which can improve prediction accuracy.


    4.Avoids Misleading Assumptions
    Imputed values are estimates, not real data. Without a flag, the system treats them as actual values, which can bias analysis.


    5.Supports Better Data Auditing
    It helps in tracking data quality issues and understanding how much data was originally missing.



    Conclusion:
    Flagging missing values before imputation ensures transparency, preserves useful information, and leads to more reliable analytics in ETL workflows.

#7.Consider a scenario where income is missing for many customers.How can this missingness itself provide business insights?
    -> Missing income data can itself be valuable because it may reveal patterns about customer behavior, data collection issues, or specific business segments.

    Explanation:

    1.Customer Behavior Insight-
    Customers who do not provide income information may be more privacy-conscious or less engaged. This can indicate a distinct customer segment.

    2.Segmentation Opportunity-
    Missing income can define a separate group (e.g., “unknown income customers”), which can be analyzed differently for marketing or targeting strategies.

    3.Potential Bias Detection-
    If income is missing more frequently in certain regions, age groups, or customer types, it may indicate bias in data collection or system design.

    4.Business Process Issues-
    High missingness may highlight problems in forms, onboarding processes, or data entry systems (e.g., optional fields being skipped).

    5.Risk and Decision-Making-
    In finance or credit scoring, customers with missing income might represent higher uncertainty or risk, influencing lending or approval decisions.

    Conclusion:
    Missing income is not just a data problem—it can provide meaningful insights into customer segments, system gaps, and business risks, helping organizations make better strategic decisions.
