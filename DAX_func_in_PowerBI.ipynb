#1  Create a calculated column TotalPrice using Quantity and Unit Price.
    -> TotalPrice = amazon_store_sales[Quantity] * amazon_store_sales[UnitPrice]

    Purpose :
    Calculates total price before discount.
#2 Create a calculated column DiscountedPrice.
    -> DiscountedPrice = amazon_store_sales[TotalPrice] *
    (1 - amazon_store_sales[Discount])

    Purpose :
    Calculates the price after applying the discount.
    Also, our dataset stores discount values as decimals (0.04, 0.29, etc.), so we'll not divide by 100.

#3 Create a measure Average Order Value (AOV).
    -> Average Order Value = DIVIDE(
    SUM(amazon_store_sales[SalesAmount]),
    DISTINCTCOUNT(amazon_store_sales[OrderID]))

    Purpose:
    Shows the average revenue generated per order.
#4 Using DAX, create a measure that calculates cumulative Year-to-Date (YTD) Sales based on the Order Date.

    Then create YTD Sales:
    YTD Sales = TOTALYTD([Total Sales],amazon_store_sales[OrderDate])

    Purpose:
