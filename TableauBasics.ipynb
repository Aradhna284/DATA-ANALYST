#Q1. Explain what Tableau is and why it is preferred over traditional spreadsheet tools for data analysis.
#Also mention any two industries where Tableau is commonly used and explain one use case.
    -> Tableau :-
    Tableau is a data visualization and BI tool that turns raw data into interactive dashboards and charts without heavy coding. You connect data, drag-drop fields, and it creates visuals instantly.
    It is preferred over Excel/Sheets  Handles big data because:
    Excel slows after 1M rows. Tableau connects live to millions of rows from databases/cloud.  Interactive visuals: In Excel we make static charts. In Tableau, dashboards have filters, drill-down, tooltips. One dashboard replaces 10 Excel sheets.
    
    Industries + Use case  :-
    Retail/E-commerce: Use case – Track daily sales, inventory, and customer segments. Example: Amazon uses Tableau-like dashboards to see which product category is selling most by region.  
    Healthcare: Use case – Monitor patient wait times, bed occupancy, treatment costs. Example: Hospitals track ER performance by department to reduce wait time.

# Q3. Create a Bar Chart showing Total Sales by Category.
#Explain:
#Which field goes to Rows
#Which field goes to Columns
#Which aggregation is applied to Sales
#Hint: Drag Category and Sales into the view and observe Tableau’s default behavior.
    -> Step 1: Drag fields to shelves  
    1.Rows shelf: Drag Category field. It’s a blue Dimension, so it creates 3 separate rows for Furniture, Office Supplies, Technology.  
    2.Columns shelf: Drag Sales field. It’s a green Measure, so Tableau automatically creates bars.
    
    Step 2: Aggregation applied to Sales
    Tableau’s default aggregation for Sales is SUM. So it adds up all sales for each Category.
    Result: Bar length = Total Sales for that Category.
    Example: Technology bar will be longest because SUM(Sales) for Technology > Furniture > Office Supplies in Superstore data.
    
    Step 3: Observe default behavior
    Tableau auto-chose a horizontal bar chart because Dimension is on Rows and Measure on Columns. If you swap them, bars become vertical.

#Q7. Explain the business importance of geographical analysis.
#Then describe how you would create a Map showing Sales by State using Sample – Superstore.

#Hint:  Use State and Sales. Tableau automatically detects geographic fields.
    -> Step 1: Business importance-
    Geographical analysis helps companies see where they are performing well/poorly by location. Used for inventory planning, warehouse location, and targeted marketing.
    Example: If California has high Sales but Texas has low Sales, company can run ads in Texas or check supply issues there.
    
    Step 2: Steps to create Map showing Sales by State-
    1.New worksheet.  
    2.Rows shelf: Drag State field. Tableau recognizes it as a geographic field and auto-converts view to a map.  
    3.Marks card → Size: Drag Sales Measure to Size. States with higher sales show as bigger circles.  
    4.Marks card → Color: Drag Sales again to Color for heat map effect. Darker color = higher sales.
    
    Result: Interactive US map where California, New York show large/dark circles, and smaller states show small/light circles.

#Q8. What is the difference between a Worksheet and a Dashboard in Tableau?
#Why is a dashboard more useful for decision-makers?
    -> Step 1: Difference-
    Worksheet: A single view/chart. Example: One bar chart of Sales by Category. You build analysis here.
    Dashboard: A canvas that combines multiple worksheets, filters, and KPIs on one page. Example: A dashboard with bar chart + map + KPI cards + filters together.
    
    Step 2: Dashboard is more useful for decision-makers because:- 1.All-in-one view: CEO doesn’t have time to open 10 worksheets. Dashboard shows Sales, Profit, Region, Category together.  
    2.Interactive filters: Decision-maker can select “West” from a filter and all charts update together.  
    3.At-a-glance KPIs: Dashboards highlight only key numbers, not raw data. Faster decisions.  
    Example: Instead of showing separate worksheet for Sales trend and Profit by Category, a dashboard puts both + a Region map on one screen for quick executive review.
#Q9. What is Tableau Prep, and why is it important in real-world data projects?
#What is Tableau Prep, and why is it important in real-world data projects?
#Then explain how you would:
#Remove null values
#Remove duplicate records
#Change incorrect data types (Answer in steps, no tool access required.)
    -> Tableau Prep :-
    Tableau Prep is a data preparation tool by Tableau. It’s used to clean, shape, and combine raw data before analysis in Tableau Desktop. Think of it as “data kitchen” where you wash + cut data before cooking visuals.
    It is important in real-world projects :-
    Real data is messy - missing values, duplicates, wrong formats. If you don’t clean it, charts will be wrong. Prep gives visual, step-by-step cleaning flow so non-coders can fix data without SQL/Python. It also saves the cleaning steps so next month’s data can be cleaned in 1 click.
    
    Cleaning tasks :-
    1.Remove null values:  
    Step 1: Load data → Click on the field with nulls, like Profit.Step 2: Use “Filter” step → uncheck “Null” values.
    Step 3: Only rows with valid Profit will flow to next step.
    
    2.Remove duplicate records:  
    Step 1: After connecting data, add “Clean” step.
    Step 2: Use “Remove Duplicates” option and select key fields like Order ID + Order Date.
    Step 3: Tableau keeps only 1 record for each unique Order ID.
    
    3.Change incorrect data types:  
    Step 1: In the data grid, find the field with wrong type. Example: Order Date stored as Text.
    Step 2: Click the data type icon next to field name.
    Step 3: Change Text → Date. Tableau will convert “1/5/2023” text to actual date format for time analysis.
