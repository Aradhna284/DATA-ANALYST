#1. Explain why publishing and sharing reports is considered more important than just building dashboards in Power BI Desktop.
    -> Dashboards in Desktop are local only. Publishing + sharing makes reports accessible to teams, allows auto-refresh, security, and collaboration. Building alone has no business impact if others can’t see/use it.
    Example: You build a sales dashboard in Desktop. If you don’t publish, your manager can’t see it. After publishing to Service, manager can view on phone + data auto-refreshes daily.

#5. Differentiate between Report and Dashboard in Power BI based on structure and usage.
    -> A Report in Power BI is a multi-page document that you build in Power BI Desktop. It has a flexible structure with multiple visuals, slicers, filters, and drill-down options on each page. The usage is for deep analysis and exploration. Users can interact with data, apply filters, and see details behind the numbers.
    Example: A “Sales Performance Report” with 4 pages.
    Page 1 shows yearly sales trend, Page 2 has region-wise map, Page 3 shows product breakdown, and Page 4 has customer details. A manager can select “North Region” from a slicer and all charts update, then drill from Year → Quarter → Month to find why sales dropped in March.
    
    2.A Dashboard is a single-page, canvas-style view created only in Power BI Service by pinning tiles from reports. Its structure is fixed to one page and tiles are read-only. The usage is for high-level monitoring of KPIs at a glance. There’s no drill-down or page navigation, just quick insights.
    Example: A CEO’s “Executive Dashboard” with 6 tiles pinned from different reports. Tile 1 shows Total Revenue card, Tile 2 has live sales gauge, Tile 3 shows top 5 products bar chart. The CEO opens it on mobile every morning to check if revenue crossed the target, but cannot change filters or drill into monthly data from there.

#6.  Explain the difference between the following three sharing methods in Power BI Service:
#.Direct Report Sharing
#.Workspace Access
#.Power BI Apps
#.Also mention one use case for each.

    -> 1. Direct Report/Dashboard Sharing:-
    This lets you share a specific report or dashboard with individual users or groups by sending them a link. The user needs a Power BI Pro license if the content is in My Workspace. Permissions can be set to allow or block resharing and building new content on the dataset.
    Example + Use case: You finish a “Q3 Budget Report” and share it directly with only the CFO and Finance Manager.
    Use case: When data is sensitive and you want to limit access to 2-3 people, not the whole team. CFO gets an email with a direct link and can view it on web or mobile.
    
    2.Workspace Access / App Workspace:-
    Here you add users to an entire workspace and assign them a role like Admin, Member, Contributor, or Viewer. Everyone in that workspace can see all reports, datasets, and dashboards inside it based on their role. It’s best for team collaboration.
    Example + Use case: Create a “Marketing Team Workspace” and add all 15 marketing members as Viewers, and 3 analysts as Contributors.
    Use case: When a whole department needs access to multiple reports. New reports added to that workspace are automatically visible to all members without sharing one by one.
    
    3.Power BI Apps:-
    An App is a packaged collection of dashboards, reports, and datasets from a workspace. You publish the app and distribute it to a large audience. As the publisher, you control navigation, branding, and who can see what. End users get a clean, app-like experience without seeing the backend workspace.
    Example + Use case: Package 3 reports + 1 dashboard into “HR Analytics App” and publish it to all 2000 employees.
    Use case: When you want to distribute standardized content company-wide. Employees see only the final app with guided navigation, but they can’t edit or see unpublished workspace content.

#7. Describe the four workspace roles available in Power BI and explain how access control is managed using these roles.
    -> Workspaces in Power BI Service use role-based access control. The role you assign decides what a user can do, following the “least privilege” principle.
    
    1.Admin :-
    Full control over the workspace. Can add/remove members, change roles, update workspace settings, and delete the workspace.
    Example: The IT Data Owner is made Admin of “Company Sales Workspace”. They can add new team leads and delete old datasets.2. Member :-
    Can do almost everything an Admin can except delete the workspace or change Admin role. Can add/remove users, publish content, update apps, and share items.
    Example: A Team Lead is a Member. They can add 5 new interns as Viewers and publish the latest “Monthly Sales Report” to the workspace.
    3.Contributor :-
    Can create, edit, and delete content like reports and datasets inside the workspace. But cannot add/remove users or share content outside the workspace.
    Example: A Data Analyst is a Contributor. They can build new visuals in the “Customer Report” but cannot give access to the Sales team. Only Members/Admins can share.
    4.Viewer :-
    Read-only access. Can view and interact with reports, dashboards, and datasets, apply filters, and export data if allowed. Cannot edit or share content.
    Example: A Sales Executive is a Viewer. They can open the dashboard, filter by their region, and download a chart, but they cannot change the report design or add new users.
    Access control: You manage it by assigning the lowest role needed for the job. Viewers for consumers, Contributors for builders, Members/Admins for owners. This keeps data secure while allowing collaboration.

#8. Why are Power BI Apps considered more secure and scalable than direct sharing? Explain this from a governance and enterprise usage perspective.
    -> Power BI Apps are more secure and scalable because they give publishers full control over what end users see and how content is distributed, unlike direct sharing which is manual and messy at scale.
    From a governance perspective: Apps hide the backend workspace. Users see only the published app with curated navigation, branding, and read-only tiles. You control updates centrally. If you fix a report, the app updates for all 2000 users automatically. With direct sharing, you’d have to share each report individually and users could see the workspace structure + unpublished drafts.
    From an enterprise usage perspective: Apps are built for large audiences. You can publish to “Entire organization” or specific Azure AD groups without adding them to the workspace. This avoids license + access chaos.
    Example: A company has 5000 employees. Using direct sharing for the “Company KPI Dashboard” means adding 5000 emails one by one and risking someone resharing. Using a Power BI App, the BI team publishes it once, assigns it to the “All Employees” group, and users get a clean app link. IT retains control, and end users can’t mess with the dataset.

#Q9. Explain Row-Level Security (RLS) in Power BI.How does RLS ensure that different users see different data using the same report?
    -> Row-Level Security is a data security feature in Power BI that restricts data access for users at the row level based on their identity. You create DAX filter rules on tables in Desktop, like [Region] = USERPRINCIPALNAME() or [ManagerID] = USERNAME(). When the report is published to Service, Power BI checks who is logged in and applies that filter automatically.
    So all users open the same report file, but each sees only the rows they are allowed to see. The report design, visuals, and calculations remain identical. RLS works for both report viewing and direct dataset queries.
    
    Example: You build one “Sales Report” with data for North, South, East, West regions. You set an RLS rule: [Region] = USERPRINCIPALNAME(). Now, aradhna@company.com is mapped to “North” in the security table. When Aradhna opens the report, she sees only North region sales. When the West manager logs in, he sees only West data. One report, multiple data views, zero chance of data leakage.

#Q10.Assume an organization shares reports without using Apps, workspace roles, or RLS.Discuss any three risks or issues that may arise related to security, governance, or data misuse, and explain how Power BI best practices help prevent them.
    -> Risk 1: Data Leakage / Unauthorized Access - Security issue
    Without RLS or workspace roles, anyone with the link can see all rows of data. A sales rep could see salary or profit data meant only for management.
    Best practice: Use RLS to filter rows by user + assign Viewer role to consumers. Only Admin/Member can see full dataset. This ensures data segregation.
    
    Risk 2: Loss of Governance Control - Governance issue
    Direct sharing creates “link sprawl”. No one knows who has access to which report. If an employee leaves, you can’t revoke access centrally. Also users can duplicate or reshape reports, creating multiple “versions of truth”.
    Best practice: Use Power BI Apps + workspace roles. Apps give central distribution and version control. Admins can remove a user from workspace and access is gone instantly. One published app = one single source of truth.
    
    Risk 3: Scalability and Maintenance Nightmare - Operational issue
    Sharing 20 reports with 1000 users individually is impossible to manage. Every update means resharing links. If dataset credentials change, all 1000 users face broken reports.
    Best practice: Publish content as an App and manage access through Azure AD groups. Update the dataset once in the workspace, and all app users get refreshed data automatically. Scales to thousands of users without manual work.
